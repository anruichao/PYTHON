```python
outp = {1:0}
for xstart in range(2,100000):
    x = xstart
    count = 1
    while True:
        if x%2==0:
            x //= 2
        else:
            x *= 3
            x += 1
        if x < xstart: 
            count += outp[x]
            break
        count += 1
    outp[xstart] = count
print(max(outp.values()))
```
