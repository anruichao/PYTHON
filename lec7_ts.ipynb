```python
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
import statsmodels.api as sm
import sqlite3
```

```python

con = sqlite3.connect('wooldridge.db')
traffic = pd.read_sql('select * from traffic2',con)
traffic.head(10)
```

```python
traffic['dtotacc'] = traffic['totacc'].diff()
traffic['dunem'] = traffic['unem'].diff()
x = traffic['dunem']
bigX = sm.add_constant(pd.concat((x,traffic['beltlaw'],traffic['spdlaw']),1))[1:]
x = x[1:]
y = traffic['dtotacc'][1:]
model = sm.OLS(y,bigX).fit()
e = y-model.predict()
model.summary()
```

```python
from statsmodels.tsa.arima_model import ARIMA
T = x.shape[0]
mat = pd.DataFrame({'ndx':np.arange(T),'x':traffic['dtotacc'].dropna().values,'e':e.values})
mat.index = pd.to_datetime(mat['ndx'])
xarima = ARIMA(mat['x'],order=(1,0,0)).fit()
earima = ARIMA(mat['e'],order=(1,0,0)).fit()
mat['xo'] = xarima.resid
mat['eo'] = earima.resid
mat
```

```python
nsim = 1000
outp = np.zeros((nsim,4))
for isim in range(nsim):
    samp = np.random.choice(mat['ndx'],T+25)
    npmat = mat.values
    rsmat = npmat[samp,:]
    xrs = rsmat[:,3]
    ers = rsmat[:,4]
    for i in range(T+25):
        if i == 0: continue
        xrs[i] += xarima.params[0] + xarima.params[1]*xrs[i-1]
        ers[i] += earima.params[0] + earima.params[1]*ers[i-1]
    xrs = xrs[25:]
    ers = ers[25:]
    xmat = np.hstack((traffic['beltlaw'].values.reshape(-1,1),traffic['spdlaw'].values.reshape(-1,1)))
    xmat = np.hstack((xrs.reshape(-1,1),xmat[1:]))
    xmat = sm.add_constant(xmat)
    ymat = xmat.dot(model.params)+ers
    mdlrs = sm.OLS(ymat,xmat).fit()
    outp[isim,:] = mdlrs.params
outp
```

```python
outp[:,0].sort()
outp[:,1].sort()
outp[:,2].sort()
outp[:,3].sort()
print(outp[int(nsim*0.025),:])
print(outp[int(nsim*0.975),:])
```
